```python
# Importing necessary packages
import pandas as pd
import numpy as np
```

```python
# Task 1: Convert a dataframe into a CSV file, excluding the row index
project = pd.DataFrame({
    "ID": ['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'], 
    "Project": ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7', 'Project 8', 'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14'], 
    "Cost": [1002000, 2000000, 4500000, 5500000, None, 680000, 400000, 350000, None, 300000, 2000000, 1000000, 3000000, 200000], 
    "Status": ['Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished', 'Failed', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Failed', 
               'Ongoing', 'Finished', 'Finished']
})
project.to_csv("Project_Data.csv", index=False)

employee = pd.DataFrame({
    "ID": ['A001', 'A002', 'A003', 'A004', 'A005'], 
    "Name": ["John Alter", "Alice Luxumberg", "Tom Sebastine", "Nina Adgra", "Amy Johny"], 
    "Gender": ['M', 'F', 'M', 'F', 'F'], 
    "City": ["Paris", "London", "Berlin", "Newyork", "Madrid"], 
    "Age": [25, 27, 29, 31, 30]
})
employee.to_csv("Employee_Data.csv", index=False)

seniority_level = pd.DataFrame({
    "ID": ['A001', 'A002', 'A003', 'A004', 'A005'], 
    "Designation Level": [2, 2, 3, 2, 3]
})
seniority_level.to_csv("Seniority_Level_Data.csv", index=False)
```

```python
# Task 2: Fill NAN values in the Cost column with the Running Sum Average
project =pd.read_csv("Project_Data.csv")
running_sum = 0
count = 0
for i in range(len(project)):
    if not pd.isna(project.iloc[i, 2]):
        running_sum += project.iloc[i, 2]
        count += 1
        avg = running_sum / count
    else:
        project.iloc[i, 2] = avg

# # built in functions
# cumulative_sum = project['Cost'].fillna(0).cumsum()
# cumulative_count = (~project['Cost'].isna()).cumsum()
# running_avg = cumulative_sum.div(cumulative_count)
# project['Cost'] = project['Cost'].fillna(running_avg)

project
```

```python
# Task 3: Split Name into First and Last Name and drop the Name column
employee = pd.read_csv("Employee_Data.csv")
employee[['First Name', 'Last Name']] = employee['Name'].str.split(' ', expand=True)
employee = employee.drop(columns=['Name'])
employee
```

```python
# Task 4: create a Final dataframe combining all three dataframes
seniority_level = pd.read_csv("Seniority_Level_Data.csv")
Final = pd.merge(pd.merge(project, employee, on='ID', how='left'), seniority_level, on='ID', how='left')
Final
```

```python
# Task 5: Create a bonus column which would be 5% of the Cost if the Status is Finished
Final['bonus'] = np.where(Final['Status']=='Finished', Final['Cost']*0.05, 0)
Final
```

```python
# Task 6: Reduce the designation level by 1 if the Status is failed and drop all the records where designation is > 4
Final['Designation Level'] = np.where(Final['Status']=='Failed', Final['Designation Level']-1, Final['Designation Level'])
Final.drop(Final[Final['Designation Level']>4].index)
Final
```

```python
# Task 7: Add courtesy titles in the First Name column and drop the gender column
Final["First Name"] = np.where(Final["Gender"]=="F", "Mrs. "+ Final['First Name'], "Mr. "+ Final['First Name'])
Final.drop(columns="Gender", inplace=True)
Final
```

```python
# Task 8: Increase designation level by 1 if age is > 29 by using apply function
Final['Designation Level'] = Final.apply(lambda row: row['Designation Level']+1 if row['Age']>29 else row['Designation Level'], axis=1)
Final
```

```python
# Task 9: Calculate the total cost per employee by grouping records by employee id and first name
TotalProjCost = Final.groupby(['ID', 'First Name'], as_index=False)["Cost"].sum()
TotalProjCost.rename({"Cost": "Total Cost"}, axis=1, inplace=True)
TotalProjCost
```
